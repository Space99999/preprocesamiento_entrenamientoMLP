```python
import os
import numpy as np
import pandas as pd
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, precision_recall_fscore_support
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline as ImbPipeline  # IMPORTANTE: Pipeline compatible con SMOTE
from collections import Counter
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint, CSVLogger
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l1_l2
from tensorflow.keras.metrics import Precision, Recall, CategoricalAccuracy
import tensorflow as tf
import joblib

# Configuración de reproducibilidad
np.random.seed(42)
tf.random.set_seed(42)
features = [
    "Flow Duration",
    "Bwd Packet Length Max",
    "Bwd Packet Length Mean",
    "Bwd Packet Length Std",
    "Flow Packets/s",
    "Flow IAT Max",
    "Bwd IAT Total",
    "Fwd Packets/s",
    "Bwd Packets/s",
    "FIN Flag Count",
    "ECE Flag Count",
    "Bwd Segment Size Avg",
    "Subflow Bwd Bytes",
    "FWD Init Win Bytes",
    "Bwd Init Win Bytes",
    "Fwd Seg Size Min",
    "Total Length of Bwd Packet",
    "Fwd Packet Length Max",
    "Fwd Packet Length Std",
    "Flow Bytes/s",
    "Flow IAT Mean",
    "Fwd Header Length",
    "Packet Length Max",
    "Packet Length Mean",
    "Packet Length Std",
    "Packet Length Variance",
    "Average Packet Size",
    "Fwd Segment Size Avg",
    "Subflow Fwd Bytes",
    "Total TCP Flow Time"
]

# [TU CÓDIGO DE CARGA DE DATOS PERMANECE IGUAL]
ruta_carpeta = r"C:/U Octavo Semestre/tesis/Preprocesamiento de datos dataset actualizado/dataset_undersampling/"
archivos = [os.path.join(ruta_carpeta, f) for f in os.listdir(ruta_carpeta) if f.endswith(".csv")]

# Columnas a eliminar
exclude_cols = ['id', 'Flow ID', 'Src IP', 'Src Port', 'Dst IP', 'Dst Port', 'Timestamp', 'Bwd URG Flags']

print(" CARGANDO DATOS...")
# [MANTÉN TODO EL CÓDIGO DE CARGA HASTA AQUÍ...]

# [CÓDIGO DE CARGA DE DATOS - MANTENER IGUAL]
dfs = []
for archivo in archivos:
    print(f"   Cargando: {os.path.basename(archivo)}")
    df_chunk = pl.read_csv(archivo)
    
    # Eliminar columnas no deseadas si existen
    cols_to_drop = [c for c in exclude_cols if c in df_chunk.columns]
    if cols_to_drop:
        df_chunk = df_chunk.drop(cols_to_drop)
    
    # Convertir todas las columnas a float64 excepto Label y Protocol
    for c in df_chunk.columns:
        if c not in ["Label", "Protocol"]:
            df_chunk = df_chunk.with_columns(pl.col(c).cast(pl.Float64))
    
    # Normalizar etiquetas dentro del chunk
    label_mapping = {
        "Botnet Ares": "Bot",
        "Web Attack - Brute Force": "webAttacks",
        "Web Attack - XSS": "webAttacks",
        "Infiltration - Communication Victim Attacker": "Infilteration",
        "Infiltration - Dropbox Download": "Infilteration",
        "Infiltration - NMAP Portscan": "Infilteration",
    }

    df_chunk = df_chunk.with_columns([
        pl.when(pl.col("Label").str.to_lowercase() == "benign")
          .then(pl.lit("Benign"))
          .when(pl.col("Label").str.contains("Attempted"))
          .then(pl.lit("Benign"))
          .when(pl.col("Label").is_in(list(label_mapping.keys())))
          .then(
              pl.col("Label").replace_strict(
                  label_mapping, 
                  return_dtype=pl.Utf8,
                  default="Other"
              )
          )
          .otherwise(pl.col("Label"))
          .alias("Label")
    ])

    # Filtrar etiquetas con "SQL"
    df_chunk = df_chunk.filter(~pl.col("Label").str.contains("SQL"))

    # Submuestreo de 40,000 registros Benign aleatorios por archivo
    benign_mask = df_chunk.filter(pl.col("Label") == "Benign")
    other_mask = df_chunk.filter(pl.col("Label") != "Benign")

    if benign_mask.height > 40000:
        benign_mask = benign_mask.sample(n=40000, with_replacement=False, shuffle=True)

    df_chunk = pl.concat([benign_mask, other_mask], rechunk=True)
    dfs.append(df_chunk)

# Concatenar todos los CSV
df = pl.concat(dfs, rechunk=True)
print(f" Dataset cargado: {df.shape}")

print(f" Distribución final de etiquetas:")
print(df['Label'].value_counts())

# Convertir a Pandas para el análisis
df_final = df.to_pandas()

print(f" Dataset final preparado: {df.shape}")
print(f"\n Distribución de clases antes del balanceo:")
print(df_final['Label'].value_counts())
print(f"Total de muestras: {len(df_final)}")

# Separar características y etiquetas
X = df_final.drop(columns=['Label'])
y = df_final['Label']

# One-hot encoding para Protocol
protocol_dummies = pd.get_dummies(X['Protocol'], prefix='Protocol')
X = X.drop('Protocol', axis=1)
X = pd.concat([X, protocol_dummies], axis=1)

X_selected = X[features + list(protocol_dummies.columns)]

# Codificación de etiquetas
le = LabelEncoder()
y_enc = le.fit_transform(y)
class_names = le.classes_

print(f" Clases: {class_names}")

# Mostrar distribución actual sin modificar
print("\n Distribución actual (sin balanceo automático):")
unique_labels, counts = np.unique(y_enc, return_counts=True)
class_distribution = {}
for i, count in enumerate(counts):
    class_name = class_names[unique_labels[i]]
    class_distribution[class_name] = count
    print(f"  {class_name}: {count} muestras")

# ==========================================
# NUEVO: VALIDACIÓN CRUZADA CON STRATIFIEDKFOLD
# ==========================================

print("\n" + "="*80)
print(" IMPLEMENTANDO VALIDACIÓN CRUZADA ESTRATIFICADA")
print("="*80)

# 1. Primero separamos un conjunto de test final que NO tocamos hasta el final
X_develop, X_final_test, y_develop, y_final_test = train_test_split(
    X_selected, y_enc, test_size=0.2, stratify=y_enc, random_state=42
)

print(f"\nDivisión inicial:")
print(f"  Desarrollo (train+val): {len(X_develop)} muestras")
print(f"  Test final: {len(X_final_test)} muestras")

# 2. Configurar validación cruzada estratificada (5 folds)
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# 3. Crear pipeline con SMOTE para webAttacks
webattacks_idx = le.transform(['webAttacks'])[0]
target_webattacks = 2000

# Pipeline que manejará el preprocesamiento
def create_preprocessing_pipeline():
    """
    Crea un pipeline que incluye SMOTE y StandardScaler
    """
    return ImbPipeline([
        ('smote', SMOTE(
            sampling_strategy={webattacks_idx: target_webattacks},
            random_state=42,
            k_neighbors=5
        )),
        ('scaler', StandardScaler())
    ])

# Arquitectura de DNN (mantener igual)
def create_improved_model(input_dim, num_classes):
    model = Sequential([
        Dense(256, activation='relu', input_shape=(input_dim,), 
              kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)),
        BatchNormalization(),
        Dropout(0.3),
        
        Dense(128, activation='relu', 
              kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)),
        BatchNormalization(),
        Dropout(0.4),
        
        Dense(64, activation='relu', 
              kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)),
        BatchNormalization(),
        Dropout(0.3),
        
        Dense(32, activation='relu', 
              kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)),
        BatchNormalization(),
        Dropout(0.2),
        
        Dense(num_classes, activation='softmax')
    ])
    
    return model

# Definir métrica F1 personalizada
class F1Score(tf.keras.metrics.Metric):
    def __init__(self, name='f1_score', **kwargs):
        super().__init__(name=name, **kwargs)
        self.precision = Precision()
        self.recall = Recall()

    def update_state(self, y_true, y_pred, sample_weight=None):
        self.precision.update_state(y_true, y_pred, sample_weight)
        self.recall.update_state(y_true, y_pred, sample_weight)

    def result(self):
        precision = self.precision.result()
        recall = self.recall.result()
        return 2 * ((precision * recall) / (precision + recall + tf.keras.backend.epsilon()))

    def reset_state(self):
        self.precision.reset_state()
        self.recall.reset_state()

# ==========================================
# VALIDACIÓN CRUZADA - ENTRENAR 5 MODELOS
# ==========================================

print("\n Iniciando validación cruzada estratificada...")

# Almacenar métricas de todos los folds
cv_scores = []
cv_histories = []
cv_reports = []
best_models = []

# Crear carpetas
os.makedirs("modelos", exist_ok=True)
os.makedirs("logs", exist_ok=True)

for fold, (train_idx, val_idx) in enumerate(skf.split(X_develop, y_develop)):
    print(f"\n{'='*20} FOLD {fold + 1}/5 {'='*20}")
    
    # 1. Dividir datos para este fold
    X_train_fold = X_develop.iloc[train_idx].reset_index(drop=True)
    X_val_fold = X_develop.iloc[val_idx].reset_index(drop=True)
    y_train_fold = y_develop[train_idx]
    y_val_fold = y_develop[val_idx]
    
    print(f"Train: {len(X_train_fold)} | Validation: {len(X_val_fold)}")
    
    # 2. CORRECCIÓN: Aplicar SMOTE + Scaling de manera separada
    try:
        # Primero aplicar SMOTE solo en el conjunto de entrenamiento
        smote = SMOTE(
            sampling_strategy={webattacks_idx: target_webattacks},
            random_state=42,
            k_neighbors=5
        )
        X_train_smote, y_train_smote = smote.fit_resample(X_train_fold, y_train_fold)
        
        print(f"Después de SMOTE - Train: {len(X_train_smote)}")
        
        # Mostrar distribución después de SMOTE en este fold
        unique, counts = np.unique(y_train_smote, return_counts=True)
        print("Distribución en train después de SMOTE:")
        for i, count in enumerate(counts):
            print(f"  {class_names[unique[i]]}: {count} muestras")
        
        # Luego aplicar StandardScaler
        scaler = StandardScaler()
        X_train_processed = scaler.fit_transform(X_train_smote)
        X_val_processed = scaler.transform(X_val_fold)
        y_train_processed = y_train_smote
        
    except Exception as e:
        print(f"Error en SMOTE para fold {fold + 1}: {e}")
        print("Usando datos originales sin SMOTE...")
        # Si SMOTE falla, usar los datos originales
        scaler = StandardScaler()
        X_train_processed = scaler.fit_transform(X_train_fold)
        X_val_processed = scaler.transform(X_val_fold)
        y_train_processed = y_train_fold
    
    # 3. Crear modelo para este fold
    model_fold = create_improved_model(X_train_processed.shape[1], len(class_names))
    
    optimizer = Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, decay=1e-6)
    model_fold.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy', Precision(name='precision'), Recall(name='recall'), F1Score()]
    )
    
    # 4. Convertir a categorical
    y_train_cat = to_categorical(y_train_processed, num_classes=len(class_names))
    y_val_cat = to_categorical(y_val_fold, num_classes=len(class_names))
    
    # 5. Calcular pesos de clase para este fold
    class_weights = compute_class_weight(
        'balanced', 
        classes=np.unique(y_train_processed), 
        y=y_train_processed
    )
    class_weight_dict = dict(zip(np.unique(y_train_processed), class_weights))
    
    # 6. Callbacks para este fold
    callbacks = [
        EarlyStopping(
            monitor='val_loss',
            patience=10,
            restore_best_weights=True,
            verbose=0
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=5,
            min_lr=1e-7,
            verbose=0
        ),
        ModelCheckpoint(
            filepath=f'modelos/model_fold_{fold+1}_best.h5',
            monitor='val_accuracy',
            save_best_only=True,
            mode='max',
            verbose=0
        )
    ]
    
    # 7. Entrenar modelo
    print(f"Entrenando modelo fold {fold + 1}...")
    history_fold = model_fold.fit(
        X_train_processed, y_train_cat,
        validation_data=(X_val_processed, y_val_cat),
        epochs=50,  
        batch_size=256,
        callbacks=callbacks,
        class_weight=class_weight_dict,
        verbose=1  # Cambiado a 1 para ver el progreso
    )
    
    # 8. Evaluar modelo en validación
    val_results = model_fold.evaluate(X_val_processed, y_val_cat, verbose=0)
    val_loss = val_results[0]
    val_acc = val_results[1]
    val_precision = val_results[2]
    val_recall = val_results[3]
    val_f1 = val_results[4]
    
    cv_scores.append({
        'fold': fold + 1,
        'val_loss': val_loss,
        'val_accuracy': val_acc,
        'val_precision': val_precision,
        'val_recall': val_recall,
        'val_f1': val_f1
    })
    
    cv_histories.append(history_fold)
    best_models.append(model_fold)
    
    # 9. Generar reporte detallado para este fold
    y_pred_fold = np.argmax(model_fold.predict(X_val_processed, verbose=0), axis=1)
    report = classification_report(y_val_fold, y_pred_fold, target_names=class_names, output_dict=True)
    cv_reports.append(report)
    
    print(f"Fold {fold + 1} - Val Acc: {val_acc:.4f}, Val F1: {val_f1:.4f}")
# ==========================================
# ANÁLISIS DE RESULTADOS DE VALIDACIÓN CRUZADA
# ==========================================

print("\n" + "="*80)
print(" RESULTADOS DE VALIDACIÓN CRUZADA")
print("="*80)

# Convertir resultados a DataFrame para análisis
cv_df = pd.DataFrame(cv_scores)

print("\nResultados por Fold:")
print(cv_df.round(4))

print(f"\nPromedios y Desviaciones Estándar:")
for metric in ['val_accuracy', 'val_precision', 'val_recall', 'val_f1']:
    mean_val = cv_df[metric].mean()
    std_val = cv_df[metric].std()
    print(f"{metric}: {mean_val:.4f} ± {std_val:.4f}")

# ==========================================
# EVALUACIÓN FINAL EN TEST SET
# ==========================================

print("\n" + "="*80)
print(" EVALUACIÓN FINAL EN TEST SET")
print("="*80)

# Seleccionar el mejor modelo (el con mayor val_accuracy promedio)
best_fold_idx = cv_df['val_accuracy'].idxmax()
best_model = best_models[best_fold_idx]
print(f"Mejor modelo: Fold {best_fold_idx + 1} (Val Acc: {cv_df.loc[best_fold_idx, 'val_accuracy']:.4f})")

# CORRECCIÓN: Preparar datos de test final usando el mismo preprocesamiento
# Entrenar el scaler final en todos los datos de desarrollo (sin SMOTE)
final_scaler = StandardScaler()
final_scaler.fit(X_develop)
X_final_test_processed = final_scaler.transform(X_final_test)
y_final_test_cat = to_categorical(y_final_test, num_classes=len(class_names))

# Evaluación final
final_results = best_model.evaluate(X_final_test_processed, y_final_test_cat, verbose=0)
final_loss = final_results[0]
final_acc = final_results[1]
final_precision = final_results[2]
final_recall = final_results[3]
final_f1 = final_results[4]

print(f"\nMétricas en Test Final:")
print(f"  Test Loss: {final_loss:.4f}")
print(f"  Test Accuracy: {final_acc:.4f}")
print(f"  Test Precision: {final_precision:.4f}")
print(f"  Test Recall: {final_recall:.4f}")
print(f"  Test F1-Score: {final_f1:.4f}")

# Predicciones finales
y_final_pred_probs = best_model.predict(X_final_test_processed, verbose=0)
y_final_pred = np.argmax(y_final_pred_probs, axis=1)

# Reporte de clasificación final
print("\nReporte de Clasificación Final:")
print(classification_report(y_final_test, y_final_pred, target_names=class_names, digits=4))

# Matriz de confusión final
cm_final = confusion_matrix(y_final_test, y_final_pred)
print(f"\nMatriz de Confusión Final:")
print(cm_final)

# ==========================================
# GRÁFICAS ESPECÍFICAS DEL MEJOR FOLD
# ==========================================

print("\n" + "="*80)
print(" GENERANDO GRÁFICAS DEL MEJOR FOLD")
print("="*80)

# Crear carpeta para las gráficas del mejor fold
os.makedirs("graficaMejorFold", exist_ok=True)

# Obtener datos del mejor fold
best_fold_idx = cv_df['val_accuracy'].idxmax()
best_model = best_models[best_fold_idx]
best_history = cv_histories[best_fold_idx]

print(f"Generando gráficas del Fold {best_fold_idx + 1} (Mejor modelo)")
print(f"Val Accuracy del mejor fold: {cv_df.loc[best_fold_idx, 'val_accuracy']:.4f}")

# Necesitamos las predicciones del mejor fold en el conjunto de validación
# Reconstruir los datos de validación del mejor fold
fold_counter = 0
for train_idx, val_idx in skf.split(X_develop, y_develop):
    if fold_counter == best_fold_idx:
        X_val_best_fold = X_develop.iloc[val_idx].reset_index(drop=True)
        y_val_best_fold = y_develop[val_idx]
        break
    fold_counter += 1

# Aplicar el mismo preprocesamiento que se usó en el mejor fold
final_scaler = StandardScaler()
final_scaler.fit(X_develop)  # Usar todos los datos de desarrollo para el scaler
X_val_best_processed = final_scaler.transform(X_val_best_fold)

# Hacer predicciones con el mejor modelo
y_pred_probs_best = best_model.predict(X_val_best_processed, verbose=0)
y_pred_best = np.argmax(y_pred_probs_best, axis=1)
y_true = y_val_best_fold

# Calcular métricas por clase para el mejor fold
from sklearn.metrics import precision_recall_fscore_support
precision_per_class, recall_per_class, f1_per_class, _ = precision_recall_fscore_support(
    y_true, y_pred_best, average=None, labels=range(len(class_names))
)

# Matriz de confusión del mejor fold
cm = confusion_matrix(y_true, y_pred_best)
cm_norm = confusion_matrix(y_true, y_pred_best, normalize='true')

print(f"\nMatriz de Confusión del Mejor Fold:")
print(cm)

# ==========================================
# MOSTRAR GRÁFICAS EN PANTALLA (GRID 2x3)
# ==========================================

plt.figure(figsize=(15, 12))

# Subplot 1: Matriz de confusión
plt.subplot(2, 3, 1)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=class_names, yticklabels=class_names)
plt.title('Matriz de Confusión')
plt.xlabel('Predicción')
plt.ylabel('Real')

# Subplot 2: Matriz de confusión normalizada
plt.subplot(2, 3, 2)
sns.heatmap(cm_norm, annot=True, fmt='.4f', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('Matriz de Confusión Normalizada')
plt.xlabel('Predicción')
plt.ylabel('Real')

# Subplot 3: Evolución del loss
plt.subplot(2, 3, 3)
plt.plot(best_history.history['loss'], label='Training Loss')
plt.plot(best_history.history['val_loss'], label='Validation Loss')
plt.title('Evolución del Loss')
plt.xlabel('Épocas')
plt.ylabel('Loss')
plt.legend()

# Subplot 4: Evolución de la precisión
plt.subplot(2, 3, 4)
plt.plot(best_history.history['accuracy'], label='Training Accuracy')
plt.plot(best_history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Evolución de la Precisión')
plt.xlabel('Épocas')
plt.ylabel('Accuracy')
plt.legend()

# Subplot 5: Distribución de predicciones por clase
plt.subplot(2, 3, 5)
pred_counts = [np.sum(y_pred_best == i) for i in range(len(class_names))]
plt.bar(class_names, pred_counts, alpha=0.7)
plt.title('Distribución de Predicciones')
plt.xlabel('Clases')
plt.ylabel('Número de Predicciones')
plt.xticks(rotation=45)

# Subplot 6: Comparación de métricas por clase
plt.subplot(2, 3, 6)
metrics_comparison = pd.DataFrame({
    'Precision': precision_per_class,
    'Recall': recall_per_class,
    'F1-Score': f1_per_class
}, index=class_names)
metrics_comparison.plot(kind='bar', ax=plt.gca())
plt.title('Métricas por Clase')
plt.xlabel('Clases')
plt.ylabel('Valor de Métrica')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()

# ==========================================
# GUARDAR GRÁFICAS INDIVIDUALES
# ==========================================

print("\nGuardando gráficas individuales...")

# Nombres de archivos
fig_names = [
    'matriz_confusion.png',
    'matriz_confusion_normalizada.png', 
    'evolucion_loss.png',
    'evolucion_precision.png',
    'distribucion_predicciones.png',
    'metricas_por_clase.png'
]

# 1. Matriz de confusión
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=class_names, yticklabels=class_names)
plt.title('Matriz de Confusión - Mejor Fold')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.tight_layout()
plt.savefig('graficaMejorFold/' + fig_names[0], dpi=300, bbox_inches='tight')
plt.close()

# 2. Matriz de confusión normalizada
plt.figure(figsize=(8, 6))
sns.heatmap(cm_norm, annot=True, fmt='.4f', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.title('Matriz de Confusión Normalizada - Mejor Fold')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.tight_layout()
plt.savefig('graficaMejorFold/' + fig_names[1], dpi=300, bbox_inches='tight')
plt.close()

# 3. Evolución del loss
plt.figure(figsize=(8, 6))
plt.plot(best_history.history['loss'], label='Training Loss')
plt.plot(best_history.history['val_loss'], label='Validation Loss')
plt.title('Evolución del Loss - Mejor Fold')
plt.xlabel('Épocas')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.savefig('graficaMejorFold/' + fig_names[2], dpi=300, bbox_inches='tight')
plt.close()

# 4. Evolución de la precisión
plt.figure(figsize=(8, 6))
plt.plot(best_history.history['accuracy'], label='Training Accuracy')
plt.plot(best_history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Evolución de la Precisión - Mejor Fold')
plt.xlabel('Épocas')
plt.ylabel('Accuracy')
plt.legend()
plt.tight_layout()
plt.savefig('graficaMejorFold/' + fig_names[3], dpi=300, bbox_inches='tight')
plt.close()

# 5. Distribución de predicciones
plt.figure(figsize=(8, 6))
pred_counts = [np.sum(y_pred_best == i) for i in range(len(class_names))]
plt.bar(class_names, pred_counts, alpha=0.7)
plt.title('Distribución de Predicciones - Mejor Fold')
plt.xlabel('Clases')
plt.ylabel('Número de Predicciones')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('graficaMejorFold/' + fig_names[4], dpi=300, bbox_inches='tight')
plt.close()

# 6. Métricas por clase
plt.figure(figsize=(8, 6))
metrics_comparison = pd.DataFrame({
    'Precision': precision_per_class,
    'Recall': recall_per_class,
    'F1-Score': f1_per_class
}, index=class_names)
metrics_comparison.plot(kind='bar')
plt.title('Métricas por Clase - Mejor Fold')
plt.xlabel('Clases')
plt.ylabel('Valor de Métrica')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.savefig('graficaMejorFold/' + fig_names[5], dpi=300, bbox_inches='tight')
plt.close()

# ==========================================
# CURVA ROC (SOLO GUARDAR, NO MOSTRAR)
# ==========================================

from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt

# Binarizar las etiquetas para ROC multiclase
y_true_bin = label_binarize(y_true, classes=range(len(class_names)))
n_classes = len(class_names)

# Calcular ROC para cada clase
fpr = dict()
tpr = dict()
roc_auc = dict()

plt.figure(figsize=(10, 8))

# Colores para cada clase
colors = ['blue', 'red', 'green', 'orange', 'purple', 'brown', 'pink', 'gray']

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_true_bin[:, i], y_pred_probs_best[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    
    color = colors[i % len(colors)]
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f'{class_names[i]} (AUC = {roc_auc[i]:.3f})')

# Línea diagonal de referencia
plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random Classifier')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Curva ROC Multiclase - Mejor Fold')
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('graficaMejorFold/curva_roc.png', dpi=300, bbox_inches='tight')
plt.close()  # No mostrar, solo guardar

print(f"\nGráficas del mejor fold guardadas en 'graficaMejorFold/':")
for i, name in enumerate(fig_names):
    print(f"  - {name}")
print(f"  - curva_roc.png")

print(f"\nResumen del mejor fold (Fold {best_fold_idx + 1}):")
print(f"  Validation Accuracy: {cv_df.loc[best_fold_idx, 'val_accuracy']:.4f}")
print(f"  Validation F1-Score: {cv_df.loc[best_fold_idx, 'val_f1']:.4f}")
print(f"  Validation Precision: {cv_df.loc[best_fold_idx, 'val_precision']:.4f}")
print(f"  Validation Recall: {cv_df.loc[best_fold_idx, 'val_recall']:.4f}")

# Mostrar métricas por clase del mejor fold
print(f"\nMétricas por clase del mejor fold:")
for i, class_name in enumerate(class_names):
    print(f"  {class_name}:")
    print(f"    Precision: {precision_per_class[i]:.4f}")
    print(f"    Recall: {recall_per_class[i]:.4f}")
    print(f"    F1-Score: {f1_per_class[i]:.4f}")
    print(f"    AUC: {roc_auc[i]:.4f}")

print(f"\n{'='*80}")
print(f" RESUMEN FINAL")
print(f"{'='*80}")
print(f" Validación Cruzada (5 folds):")
print(f"   Accuracy: {cv_df['val_accuracy'].mean():.4f} ± {cv_df['val_accuracy'].std():.4f}")
print(f"   F1-Score: {cv_df['val_f1'].mean():.4f} ± {cv_df['val_f1'].std():.4f}")
print(f" Test Final:")
print(f"   Accuracy: {final_acc:.4f}")
print(f"   F1-Score: {final_f1:.4f}")
print(f"{'='*80}")
```
